## Anordning

```python
urlAnordning= "http://nvpub.vic-metria.nu/friluftsliv/rest/anordning"
header ="accept: text/plain;charset=utf-8"
import urllib3, json
import pandas as pd 

http = urllib3.PoolManager()
r = http.request(
    'GET',
    urlAnordning,
    headers={
        "accept": "application/json"
    })

print(r.status)
data = json.loads(r.data.decode('utf-8')) 
#print(type(data))
print(r.info()) 
```

```python
from flatten_json import flatten
dftotAnordning = pd.json_normalize(data)  
dftotAnordning.sample(10)
```

```python
dftotAnordning["anordning.kategori"].value_counts(ascending=True)
```

# Led
